#### How many movies are there that are both in the "Thriller" genre and contain the word love anywhere in the keywords?

```python
%%sql 
SELECT
    COUNT(DISTINCT m.title ) AS count
    
FROM
    movies AS m
JOIN
genremap AS g
ON
g.movie_id = m.movie_id
JOIN
genres AS gr
ON
gr.genre_id = g.genre_id
JOIN
keywordmap kw
ON
kw.movie_id =m.movie_id
JOIN
keywords k
ON
kw.keyword_id = k.keyword_id
WHERE
(genre_name = "Thriller"  and keyword_name LIKE '%love%')
;
```

#### How many movies are there that were released between 1 August 2006 ('2006-08-01') and 1 October 2009 ('2009-10-01') that have a popularity score of more than 40 and a budget of less than 50 000 000?

```python
%%sql
SELECT
    COUNT(*) AS m_count,
    popularity,
    budget,
    title,
    release_date
FROM
    movies
WHERE 
    release_date  BETWEEN ('2006-08-01') and ('2009-10-01')
    AND budget < 50000000
    AND popularity >40
      
;
```

#### How many unique characters has "Vin Diesel" played so far in the database?

```python
%%sql
SELECT
a.actor_name,
COUNT (DISTINCT c.characters)
FROM
  actors AS a
JOIN
casts AS c
ON
a.actor_id = c.actor_id
WHERE 
    actor_name = 'Vin Diesel'
;
```

#### what are the genres of the movie The Royal Tenenbaums?

```python
%%sql
SELECT 
m.movie_id,
m.title,
gr.genre_name
FROM
    movies AS m
JOIN
genremap AS g
ON
g.movie_id = m.movie_id
JOIN
genres AS gr
ON
gr.genre_id = g.genre_id
WHERE
m.title = 'The Royal Tenenbaums'
;
```

#### What are the three production companies that have the highest movie popularity score on average, as recorded within the database?

```python
%%sql
SELECT
m.movie_id,
AVG(m.popularity) AS avg_popurality,
pc.production_company_name
FROM
movies AS m
JOIN
productioncompanymap AS pm
ON
m.movie_id = pm.movie_id
JOIN
productioncompanies AS pc
ON
pc.production_company_id = pm.production_company_id
GROUP BY
Production_company_name
ORDER BY
avg_popurality DESC
LIMIT 3;
```

#### How many female actors (i.e. gender = 1) have a name that starts with the letter "N"?

```python
%%sql
SELECT
COUNT(*)
FROM
actors
WHERE
gender =1
AND 
actor_name LIKE 'N%';
```

#### Which genre has, on average, the lowest movie popularity score?

```python
%%sql
SELECT 
m.movie_id,
g.genre_name,
AVG(popularity) AS avg_popularity
FROM
movies AS m
JOIN
genremap AS gm
ON
gm.movie_id = m.movie_id
JOIN
genres AS g
ON
g.genre_id = gm.genre_id
GROUP BY g.genre_name
ORDER BY avg_popularity ASC;
```

#### Which award category has the highest number of actor nominations (actors can be male or female)? (Hint: Oscars.name contains both actors' names and film names.)

```python
%%sql
SELECT
DISTINCT o.award,
    COUNT(O.name) AS nominee 
FROM
    Oscars AS o
JOIN
actors AS s
ON
o.name = s.actor_name
    GROUP BY o.award   
ORDER BY nominee  DESC
;
```

### Creating View

#### DStv will be having a special week dedicated to the actor Alan Rickman. Which of the following queries would create a new view that shows the titles, release dates, taglines, and overviews of all movies that Alan Rickman has played in?

```python
CREATE VIEW Alan_Rickman_Movies AS
SELECT
    title,
    release_date,
    tagline,
    overview
FROM
    Movies
LEFT JOIN
    Casts ON Casts.movie_id = Movies.movie_id 
LEFT JOIN
    Actors ON Casts.actor_id = Actors.actor_id 
WHERE
    Actors.actor_name = 'Alan Rickman';
```
